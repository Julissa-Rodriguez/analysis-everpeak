```python
# Limpia separadores y convierte columnas numéricas en eco
eco.columns = eco.columns.str.strip() \
                         .str.lower() \
                         .str.replace(' ', '_') \
                         .str.replace('/', '_') \
                         .str.replace('(', '', regex=False) \
                         .str.replace(')', '', regex=False) \
                         .str.replace('%', 'pct') \
                         .str.replace('-', '_')  # si hubiera guiones

print("Columnas de eco después de limpieza:")
print(eco.columns)

# eco['city_gdp_capita'] = ....astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
eco['city_gdp_capita'] = eco['city_gdp_capita'].astype(str) \
                                   .str.replace('.', '', regex=False) \
                                   .str.replace(',', '.', regex=False) \
                                   .astype(float)

# eco['unemployment_pct'] = ...

eco['unemployment_pct'] = eco['unemployment_pct'].astype(str) \
                                         .str.replace('%', '', regex=False) \
                                         .str.replace(',', '.', regex=False) \
                                         .astype(float)

# eco['population_m'] = ...
eco['population_m'] = eco['population_m'].astype(str) \
                               .str.replace(',', '.', regex=False) \
                               .astype(float)

# Calcula la población total en unidades absolutas (Multiplica * 1000000)
# eco['population'] = ...
eco['population'] = eco['population_m'] * 1_000_000

# verificar el cambio
eco.info()
eco.head(3)
```

### 5.1 Unir tráfico (tabla principal) con indicadores económicos

- Une ambos DataFrames y define como **claves de unión** a `city` y `year`.

<details>

Aplica una unión de tipo "inner" para mantener las ciudades y años presentes en ambos datasets.

```python
print(traffic_city_year_2024.columns.tolist())
print(eco_2024.columns.tolist())
```
